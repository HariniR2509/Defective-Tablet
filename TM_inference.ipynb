```python
#numeric values will be displayed in a more familiar format.
np.set_printoptions(suppress=True)


model = load_model("keras_model.h5", compile=False)

class_names = open("labels.txt", "r").readlines()

data = np.ndarray(shape=(1, 224, 224, 3), dtype=np.float32)

image = Image.open("mint/test/contamination/pill_mint_contamination_007.png").convert("RGB")

#fit
size = (224, 224)
image = ImageOps.fit(image, size, Image.Resampling.LANCZOS)       #Lanczos resampling- resizing images.
#The image is resized to fit within the bounding box and if any excess parts are cropped from the center.

image_array = np.asarray(image)

normalized_image_array = (image_array.astype(np.float32) / 127.5) - 1

data[0] = normalized_image_array

prediction = model.predict(data)
index = np.argmax(prediction)       # index of the maximum probability in the prediction array. This index corresponds to the predicted class.
class_name = class_names[index]
confidence_score = prediction[0][index]


print("Class:", class_name[2:], end="")
print("Confidence Score:", confidence_score)
```
